```python
def sample_noise(size):
    z = torch.FloatTensor(size, 100)
    z.data.normal_()  # generating latent space based on normal distribution
    return z

import torch

window_length = 65536 
model_capacity_size = 32  
use_batchnorm=False
num_channels = 1

noise = sample_noise(2)
noise.shape

from models import WaveGANGenerator

generator = WaveGANGenerator(
            slice_len=window_length,
            model_size=model_capacity_size,
            use_batch_norm=use_batchnorm,
            num_channels=num_channels,
        )
```

```python
def load_model(type_):
    '''
    type_: male or female voice to generate
    '''
    if type_=='male':
        checkpoint = torch.load('./gan_male.tar')
    else:
        checkpoint = torch.load('./gan_female1.tar')
    return checkpoint
```

```python
checkpoint = load_model('male')

generator.load_state_dict(checkpoint["generator"])

x =generator(noise)
```
